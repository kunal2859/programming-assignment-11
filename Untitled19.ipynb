```python
def uncommon(s1,s2):
    list_s1=s1.split()
    list_s2=s2.split()
    uc_words=""
    for i in list_s1:
        if i not in list_s2:
            uc_words=uc_words+" "+i
    for j in list_s2:
        if j not in list_s1:
            uc_words=uc_words+" "+j
    return uc_words
a="Kunal Aggarwal"
b="Ishank Aggarwal"
print(uncommon(a,b))
```

```python
def dup_char(str):
    x=[]
    for i in str:
        if i not in x and str.count(i)>1:
            x.append(i)
    print(" ".join(x))
str="Kunal Aggarwal"
dup_char(str)
```
